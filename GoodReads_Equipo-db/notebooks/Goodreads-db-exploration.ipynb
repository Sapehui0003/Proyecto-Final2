## Genres

```python
# Obtenemos la columna de genres
series_genres = df_books['genres']
series_genres
```

```python
# Borramos posibles valores nan
series_genres = series_genres.dropna()
```

```python
list_genres = series_genres.str.split(r"[;,]", regex=True).values.tolist()
                                    # Se identificó que los generos vienen separados tanto por , como por ;
list_genres
```

```python
flat_list_genres = [x2 for x1 in list_genres for x2 in x1] # list comprehension
flat_list_genres
```

```python
pd.Series(flat_list_genres).unique()
```
